```python
import numpy as np
import pandas as pd
```

```python
labels=['a','b','c']
my_list=[1,2,3]
arr=np.array([10,20,30])
d={1:10,2:20,3:30}
```

```python
pd.Series(my_list) #this is how we can make any data type into series
```

```python
pd.Series(my_list,index=labels) # this is how you can modifiy any index of a series
```

```python
pd.Series(arr,index=labels)
```
