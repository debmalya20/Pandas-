```python
import numpy as np
import pandas as pd
```

```python
employes=pd.DataFrame({
    'employes_id':[1,2,3,4,5],
    'name':['john','anna','peter','linda','bob'],
    'depertment':['hr','it','finance','it','hr']
})
#depertment 2 salary info
salaryes=pd.DataFrame({
    'employes_id':[1,2,3,6,7],
    'salary':[60000,34000,23000,230000,23400],
    'bonus':[2000,3000,2000,3000,2000]
})
```

```python
employes
```

```python
salaryes
```

```python
pd.merge(employes,salaryes,on='employes_id',how='inner')
```

```python
pd.merge(employes,salaryes,on='employes_id',how='outer')
```

```python
pd.merge(employes,salaryes,how='left')
```

```python
pd.merge(employes,salaryes,how='right')
```

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'A5'],
    'C': ['C3', 'C4', 'A5']
})
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```
