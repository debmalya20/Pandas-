```python
dl=pd.read_csv(r"anime.csv")
```

```python
def extract_txt(txt):
    data=" "
    check=False
    for i in txt:
        if i == ")":
            break
        if check == True:
            data+=i
        if i =="(":
            check=True
    return data
            
```

```python
dl["Eppisodes"]=dl["Title"].apply(extract_txt)
```

```python
dl
```

```python
dl['Eppisodes']=dl["Eppisodes"].str.replace("eps","")
```

```python
dl
```

```python
dl['Eppisodes']=dl['Eppisodes'].astype(int)
```

```python
dl
```

```python
dl["Eppisodes"].dtype
```

```python
def date_extract(txt):
    data=""
    for i in range(len(txt)):
        if txt[i]==")":
            for j in range(i+1,i+20):
                data+=txt[j]
    return data
    
```

```python
dl["Date"]=dl["Title"].apply(date_extract)
```

```python
dl
```

```python
from dateutil.relativedelta import relativedelta
from datetime import datetime

def calculate_total_months(period):
    try:
        start_str, end_str = period.split(' - ')
        start_date = datetime.strptime(start_str, '%b %Y')
        end_date = datetime.strptime(end_str, '%b %Y')
        r = relativedelta(end_date, start_date)
        return r.years * 12 + r.months + 1  # +1 to include the starting month
    except:
        return None

dl['Months'] = dl['Date'].apply(calculate_total_months)
```
